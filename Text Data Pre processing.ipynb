```python
pos_rev1 = pd.DataFrame(pos_rev).drop(['text', 'summary'], axis=1)
neg_rev1 = pd.DataFrame(neg_rev).drop(['text', 'summary'], axis=1)
```

```python
pos_rev1_set = pos_rev1.iloc[:10000,:]
neg_rev1_set = neg_rev1.iloc[:5000,:]

train1 = pd.concat([pos_rev1_set, neg_rev1_set], axis=0, ignore_index=True)
train1.rename(columns = {'rating':'label'}, inplace = True)
```
